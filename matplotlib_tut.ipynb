```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
```

```python
x = np.array([1,2,3,4,5])
```

```python
y = np.array([31,1122,23,334,64])
```

```python
#Functional Method of plotting
```

```python
plt.plot(x,y,'purple')
plt.show()
```

```python
z = y**0.5
```

```python
plt.subplot(1,5,(1,2))
plt.plot(x,y,'r--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')
plt.legend('r')
plt.subplot(1,5,(4,5))
plt.plot(x,z,'b*-')
plt.xlabel('x')
plt.ylabel('z')
plt.title('title')
plt.legend('b')
plt.show()
```

```python
# OOP method
```

```python
fig = plt.figure()

axes = fig.add_axes([0.2,0.1,0.8,0.8])
axes.plot(x,y,'b*-')
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')
axes.set_title('Plot')
axes.legend('r')

axes2 = fig.add_axes([0.6,0.6,0.2,0.2])
axes2.set_xlabel('')
axes2.plot(x,y**0.5,'r')
axes2.legend('b')

fig.savefig('MyPlot.png')
```

```python
#Plotting using subplots()
```

When plotting using subplots function we use 
tuple unpacking, as subplots function automatically 
sets the axes

```python
fig, axes = plt.subplots()
axes.plot(x,y,'green')
axes.set_title('Subplots plot')
axes.set_xlabel('X-axis')
axes.set_ylabel('Y-axis')

fig.savefig('Subplots.pdf')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(12,3),dpi=100)
for i,ax in enumerate(axes):
    ax.plot(x,y**(i*1.5+2),'green',label='plot'+' '+str(i+1))
    ax.set_title('Subplots plot')
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.legend(loc=0)
fig
plt.tight_layout()
plt.savefig('Subplot_2.pdf')
```

```python
x1 = np.linspace(0,20,21)
```

```python
fig, axes = plt.subplots(figsize = (10,6),dpi=100)
axes.plot(x1,x1+3,x1,2*x1-6,lw=3,ls='-',marker ='o', markerFaceColor = "red", markersize = 8)
axes.set_xlim([0,15])
axes.set_ylim([-5,25])
axes.set_title('Intersection of two lines generated using Matplotlib\ny = x+3\ny = 2x-6')
axes.set_xlabel('---------- X-values-------->')
axes.set_ylabel('---------- Y-values-------->')
axes.legend('bo',loc=4)
plt.savefig('two_lines.pdf')
```
