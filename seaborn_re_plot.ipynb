```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Tips-SBS.xlsx')
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex',markers=['o','*'])
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex',markers=['o','*'],scatter_kws={'s':70})
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex',col='day',row='time',aspect=0.6,height=8)
```
