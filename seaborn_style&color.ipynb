```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_excel('Tips-SBS.xlsx')
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=df)
sns.despine(bottom=True)
```

```python
plt.figure(figsize=(50,20))
sns.countplot(x='tip',data=df,hue='sex')
```

```python
sns.set_style('darkgrid')
#sns.set_context(context='poster',font_scale=1)
sns.countplot(x='time',data=df,hue='sex',palette='magma')
```

```python
sns.lmplot(x='total_bill',y='tip',data=df,hue='sex',palette='seismic',aspect=0.8,height=8,markers=['o','*'])
```
