```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
poke = pd.read_csv('Pokemon.csv')
```

```python
poke.head()
```

```python
iris = pd.DataFrame()
```

```python
iris['Name'] = poke['Name']
```

```python
iris['Total'] = poke['Total']
iris['HP'] = poke['HP']
iris['Attack'] = poke['Attack']
iris['Defense'] = poke['Defense']
iris['Speed'] = poke['Speed']
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
pg = sns.PairGrid(iris)
pg.map_diag(sns.kdeplot)
pg.map_lower(plt.scatter)
pg.map_upper(sns.kdeplot)
```
