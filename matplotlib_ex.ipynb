```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
ax1.plot(x,y,'r')
ax2.plot(x,y,'r')
plt.tight_layout()
```

```python
x = np.linspace(0,25,26)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
ax2.set_title('zoom')
ax1.plot(x,z)
ax2.plot(x,y)
ax1.set_xlim([0,100])
ax1.set_ylim([0,10000])
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
fig
plt.tight_layout()
```

```python
fig, ax = plt.subplots(1,2)
ax[0].plot(x,y,'b--')
ax[1].plot(x,z,'r',lw=3)
fig
plt.tight_layout()
```

```python
fig, ax = plt.subplots(1,2,figsize = (12,3))
ax[0].plot(x,y,'b--')
ax[1].plot(x,z,'r--',lw=3)
fig
plt.tight_layout()
```
